```python
import tkinter as tk
from tkinter import messagebox

# Correct credentials
USERNAME = "admin"
PASSWORD = "password123"

# Function to handle login
def handle_login():
    username = username_entry.get()
    password = password_entry.get()
    if username == USERNAME and password == PASSWORD:
        login_success()
    else:
        messagebox.showerror("Error", "Invalid username or password")

# Function to show success message
def login_success():
    # Create a new window
    success_window = tk.Toplevel(root)
    success_window.title("Welcome")
    success_window.geometry("400x200")
    root.attributes('-fullscreen', True)

    # Display message
    message = tk.Label(success_window, text="Wassup", font=("Arial", 24))
    message.pack(expand=True)

    # Close the main login window
    root.withdraw()

# Create the main application window
root = tk.Tk()
root.title("Login Page")
root.attributes('-fullscreen', True)  # Make the window fullscreen

# Centered frame for login form
frame = tk.Frame(root)
frame.pack(expand=True)

# Username label and entry
username_label = tk.Label(frame, text="Username:", font=("Arial", 16))
username_label.grid(row=0, column=0, padx=10, pady=10, sticky="e")
username_entry = tk.Entry(frame, font=("Arial", 16))
username_entry.grid(row=0, column=1, padx=10, pady=10)

# Password label and entry
password_label = tk.Label(frame, text="Password:", font=("Arial", 16))
password_label.grid(row=1, column=0, padx=10, pady=10, sticky="e")
password_entry = tk.Entry(frame, font=("Arial", 16), show="*")
password_entry.grid(row=1, column=1, padx=10, pady=10)

# Login button
login_button = tk.Button(frame, text="Login", font=("Arial", 16), command=handle_login)
login_button.grid(row=2, column=0, columnspan=2, pady=20)

# Exit button
exit_button = tk.Button(root, text="Exit", font=("Arial", 16), command=root.destroy)
exit_button.pack(side="bottom", pady=10)

# Run the application
root.mainloop()
```
